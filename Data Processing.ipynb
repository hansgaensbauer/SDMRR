```python
import numpy as np
import scipy.signal as sg
import matplotlib.pyplot as plt
```

```python
t2data = np.load('t2data.npy', allow_pickle=True)
t2times = np.load('t2times.npy', allow_pickle=True)

toremove = [14,24,47,55,62,67,77]

t2data = np.delete(t2data, toremove)
t2times = np.delete(t2times, toremove)
```

```python
%matplotlib notebook
plt.figure(figsize = (9,5))
plt.plot(t2data)
plt.show()
```

```python
plt.figure(figsize = (9,5))
b,a = sg.butter(2, 0.15)
z = sg.filtfilt(b, a, t2data)
plt.plot(t2times, t2data)
plt.title("Bioreactor T2")
plt.xlabel("Timestamp")
plt.ylabel("T2 (s)")
plt.plot(t2times, z)
plt.show()
```
